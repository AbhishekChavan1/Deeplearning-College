```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import os
```

```python
Dataset = pd.read_csv('/kaggle/input/bank-customer-churn-prediction/Churn_Modelling.csv')
Dataset.head()
```

```python
Dataset.info()
```

```python
x = Dataset.drop(['RowNumber','CustomerId','Surname','Exited'],axis=1)
x.head(2)
```

```python
y = Dataset.Exited.values
y
```

```python
x = pd.get_dummies(x,columns=['Geography','Gender'],drop_first=True)
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
len(x_train)
len(x_test)
```

```python
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
model = DecisionTreeClassifier(max_depth=3)
model.fit(x_train_sc,y_train)
```

```python
y_pred = model.predict(x_test_sc)
```

```python
y_pred
```

```python
y_test
```

```python
print(f'The Accuracy :- {accuracy_score(y_pred,y_test)*100}')
print(f'The Precision score : - {precision_score(y_pred,y_test)*100}')
```

```python
fig = plt.figure(figsize=(15,10))
tree_plot = plot_tree(model,feature_names=x.columns,precision=5,label='all',filled=True)
```
