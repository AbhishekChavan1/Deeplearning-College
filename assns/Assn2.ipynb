```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel('sales_data.csv.xlsx')
df.head()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.title('Sales Distribution Across Product Categories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Relationship between Sales and Profit')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Quantity'], bins=5, edgecolor='black')
plt.xlabel('Quantity Sold')
plt.ylabel('Frequency')
plt.title('Distribution of Quantity Sold')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Discount', data=df)
plt.title('Comparison of Discounts across Regions')
plt.xlabel('Region')
plt.ylabel('Discount')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
selected_features = ['Sales', 'Quantity', 'Profit', 'Discount']
correlation_matrix = df[selected_features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
